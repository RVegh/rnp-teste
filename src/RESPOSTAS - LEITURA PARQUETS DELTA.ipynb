O Databricks community possui limitações no que tange o download de arquivos no DBFS, logo, optei por evidenciar com imagens no README que todos os parquets foram criados e salvos em formato delta e todos possuem o diretório de delta_logs.

Abaixo, segue a leitura destes parquets via spark diretamente do DBFS para melhor visualização das respostas.

```
from pyspark.sql import SparkSession

#Cria a sessão Spark
spark = SparkSession.builder \
    .appName('Convert to PySpark DataFrame') \
    .config('spark.sql.legacy.pathOptionBehavior.enabled', True) \
    .config('spark.databricks.delta.formatCheck.enabled', False)  \
    .getOrCreate()
```

```
#PERGUNTA 1 - Qual país possui a maior quantidade de itens cancelados?
file_path = '/dbfs/Users/devrvegh@gmail.com/dados_delta/Pergunta_01_REAL1/part-00000-3c58863f-4db7-4b55-ae37-1be9d536cfa5-c000.snappy.parquet'

df = spark.read.parquet(file_path)
display(df)
```
